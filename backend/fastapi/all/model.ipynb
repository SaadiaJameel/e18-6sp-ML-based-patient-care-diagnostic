```python
import pandas as pd
from sklearn.tree import DecisionTreeClassifier
from sklearn.model_selection import train_test_split
from sklearn import metrics
```

```python
# Dataset loading and exploratory data analysis
df = pd.read_csv('formatted_dataset_all.csv')
df
```

```python
# Drop the target variable
X = df.drop(['disease','cures', 'doctor','risk_level'], axis=1)
disease = df['disease']
Cures = df['cures']
Doctor = df['doctor']
Risk = df['risk_level']
len(disease)
```

```python
#remove and assign diseases with one distinct symptom
import array as arr

df2 = pd.read_csv('formatted_dataset_all.csv')
sympotoms_direct_disease = []
sympotoms_direct_disease_dict={}

for i in range(len(symptoms_array_final)):
    arr1=arr.array('i',df2[symptoms_array_final[i]])
    if(arr1.count(1)==1):
        index = arr1.index(1)
        sympotoms_direct_disease.append(symptoms_array_final[i])
        sympotoms_direct_disease_dict[symptoms_array_final[i]] = disease[index]
        
len(sympotoms_direct_disease_dict)

X2=X
sympotoms_indirect_disease = []

for i in X2:
    if i in sympotoms_direct_disease_dict:
        X2 = X2.drop([i], axis=1)
    else:
        sympotoms_indirect_disease.append(i)

cure_and_disease={}
doctor_and_disease={}
risk_and_disease={}

#cure and doctor type recommendation
for i in range(len(disease)):
    cure_and_disease[disease[i]] = Cures[i]
    doctor_and_disease[disease[i]] = Doctor[i]
    risk_and_disease[disease[i]] = Risk[i]

risk_and_disease
```

```python
# Dataset loading and exploratory data analysis
df = pd.read_csv('formatted_dataset_all.csv')
df

prediction = ""
input = "fever"
input_array = input.split(',')
input_array_final = []

for i in range(len(input_array)):
    x = input_array[i].split(' ')
    string = input_array[i]
    if(len(x)>1):
        string = ""
        for j in range(len(x)):
            string = string+x[j]
            string = string + '_'    
        string = string[:-1]
    
    input_array_final.append(string)
    
#if sympotom in the direct symptom to disease predict
    for i in input_array_final:
        if(i in sympotoms_direct_disease_dict):
            print(sympotoms_direct_disease_dict.get(i))
    
labels=[]

symptoms_array_in_more_than_5 = df.columns
symptoms_array_in_more_than_5 = symptoms_array_in_more_than_5[:-3]

for i in range(len(symptoms_array_final)):
    if(symptoms_array_final[i] in input_array_final):
        labels.append(1)
    else:
        labels.append(0)
print(labels)
X
```

```python
# Build the decision tree classifier- gini
clf_gini= DecisionTreeClassifier(criterion='gini',
                                max_depth= 2,
                                random_state= 0)
#Train the classifier
clf_gini.fit(X, disease)

# Predicting results for the test set
clf_gini.predict([labels])
```
