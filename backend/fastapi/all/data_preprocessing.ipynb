```python
import pandas as pd
import numpy as np
```

```python
all_disease = pd.read_csv('dataset.csv')
all_disease
```

```python
diseases = []

for i in range(all_disease.shape[0]):
    if (all_disease['disease'][i] in diseases):
        print(all_disease['disease'][i])
    else:
        diseases.append(all_disease['disease'][i])

 
symptoms = all_disease['symptoms']
symptoms_array= []
symptoms_array_unique=[]
symptoms_array_final=[]


for i in range(symptoms.size):
    x = symptoms[i].split(',')
    for j in range(len(x)):
        symptoms_array.append(x[j])

for i in range(len(symptoms_array)):
    if not ((symptoms_array[i] in symptoms_array_unique)):
        symptoms_array_unique.append(symptoms_array[i])
            
for i in range(len(symptoms_array_unique)):
    x = symptoms_array_unique[i].split(' ')
    string = symptoms_array_unique[i]
    if(len(x)>1):
        string = ""
        for j in range(len(x)):
            string = string+x[j]
            string = string + '_'    
        string = string[:-1]
    
    symptoms_array_final.append(string)

with open("output.txt", "w") as txt_file:
    for line in symptoms_array_final:
        txt_file.write("'"+"".join(line)+"'"+",") # works with any number of elements in a line
        
with open("output_1.txt", "w") as txt_file:
    for line in symptoms_array_unique:
        txt_file.write("'"+"".join(line) + "',") # works with any number of elements in a line
        
with open("output_2.txt", "w") as txt_file:
    for i in range(len(symptoms_array_final)):
        txt_file.write('"'+"".join(symptoms_array_final[i]) +'":'+str(0)+",\n")
        
symptoms_array_final.append('disease')
symptoms_array_final.append('cures')
symptoms_array_final.append('doctor')

all_disease
```

```python
df = pd.DataFrame(columns=symptoms_array_final)
df.to_csv('formatted_dataset_all.csv', index=False)

row2 = np.array([])

for i in range(all_disease.shape[0]):
    row = all_disease.iloc[i].to_numpy()
    row3 = row[1].split(',')
    row2 = []
    
    for j in range(len(symptoms_array_unique)):
        if(symptoms_array_unique[j] in row3):
            row2.append(1)
        else:
            row2.append(0)
    
    row2.append(row[0])
    row2.append(row[2])
    row2.append(row[3])

    df.loc[len(df.index)] = row2 

df.to_csv('formatted_dataset_all.csv', index=False)
df
```
